Insight

Berdasarkan visualisasi data diperoleh bahwa pada sumbu x dari rentang 500-1000 , konsentrasi ruangan mencapai lebih dari 4000

```python
# Visualisasi data menggunakan bar chart untuk perbandingan banyaknya ruangan

# Logika untuk menampilkan bar chart
sumkitch= Data["kitchen"].sum()
sumliv = Data["livingroom"].sum()
sumbed = Data["bedroom"].sum()
sumoff = Data["office"].sum()
sumkitliv = Data["kitchen_livingroom"].sum()

data1 = {
    'Ruangan':["kitchen", "living room", "bedroom", "office", "kitchen_livingroom"],
    'sum' : [sumkitch, sumliv, sumbed, sumoff, sumkitliv ]
}

df = pd.DataFrame(data1)

df.plot(kind="bar", x="Ruangan", y ="sum",ylabel="Banyaknya Ruangan", title="Perbandingan banyaknya ruangan")
```

Insight

Berdasarkan visualisasi data, banyaknya ruangan yang terpakai oleh banyaknya pemilik yaitu bedroom dengan jumlah lebih 3000 ruangan, sedangkan sedikitnya ruangan yang terpakai oleh banyaknya pemilik yaitu kitchen_livingroom dengan jumlah kurang dari 1500

```python
# Visualisasi data menggunakan bar chart untuk keterisian ruangan

# Logika untuk menampilkan bar chart
a = Data.loc[Data["occupancy_ground_truth"]== 0]
b = Data.loc[Data["occupancy_ground_truth"]== 1]

counta = a["occupancy_ground_truth"].value_counts()
countb = b["occupancy_ground_truth"].value_counts()

x = counta.values
y = countb.values
banyaksng = x[0]
banyakisi = y[0]

data2 = {
    'Status': ["Terisi", "Kosong"],
    'Banyak':[banyakisi,banyaksng]
}

df = pd.DataFrame(data2)

df.plot(kind="bar", x="Status", y="Banyak",ylabel="Banyaknya Ruangan", title="Keterisian Ruangan")
```

Insight

Berdasarkan visualisasi data, keterisian ruangan lebih banyak yang kosong dengan jumlah lebih dari 7000 ruangan dan yang terisi kurang dari 3000 ruangan

```python
# Visualisasi data menggunakan line chart untuk Perbandingan waktu dengan konsentrasi CO2 dalam Ruangan

# Logika untuk menampilkan line chart

day1 = Data[:96]
day2 = Data[96:192]
day3 = Data[192:288]
day4 = Data[288:384]
day5 = Data[384:480]


meanday1 = day1["indoor_co2_concentration"].mean()
meanday2 = day2["indoor_co2_concentration"].mean()
meanday3 = day3["indoor_co2_concentration"].mean()
meanday4 = day4["indoor_co2_concentration"].mean()
meanday5 = day5["indoor_co2_concentration"].mean()


data3 = {
    'Tanggal':["30-01-23", "31-01-23", "1/2/2023", "2/2/2023", "3/2/2023"],
    'sum' : [meanday1,meanday2,meanday3,meanday4,meanday5]
}

df = pd.DataFrame(data3)

df.plot(kind="line", x="Tanggal", y ="sum",xlabel="Konsentrasi CO2 dalam Ruangan(ppm)", title="Perbandingan waktu dengan konsentrasi CO2 dalam Ruangan")
```
