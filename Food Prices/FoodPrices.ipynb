# Data Cleansing

Pada Data Food Prices, terdapat kesalahan pada kategori Unit of Measurements. Kesalahan tersebut terletak di beberapa data dengan Kategori Food Items Susu mendapatkan Unit of Measurements 1 kg, berbeda dengan data Kategori Susu lainnya yang mendapatkan Unit of Measurements nya adalah 1 L. 

Maka dari itu, kita harus bersihkan data tersebut dengan mengubah Unit of Measurements Susu menjadi sama semua yaitu 1 L.

- Memperbaiki Unit of Measurements Susu yang awalnya 1 kg menjadi 1L

```python
# Membuat looping untuk replace data
for index in range(1020, 1080):
    if fp.loc[index, 'Unit of Measurement'] == '1 Kg':
        fp.at[index, 'Unit of Measurement'] = '1 L'

# Hasil Cleansing Data
print(fp.to_string())
```

6. Korelasi

- Korelasi Nilai Harga Pangan Mata Uang USD dengan Nilai Harga Pangan Mata Uang AUD

```python
# Atur Data untuk Mata Uang AUD
a = fp.loc[(fp["Currency "]== "AUD")] 

# Mencari Koefisien Korelasi
corr= a["Average Price "].corr(a["Price in USD"])

# Buat Scatter Plot
ax = a.plot(kind="scatter", x="Price in USD", y="Average Price ")

#Atur Title dan Penamaan Sumbu
ax.set_title('Korelasi Nilai Harga Pangan dalam USD dengan Nilai Harga Pangan dalam mata uang AUD')
ax.set_xlabel('Harga (USD)')
ax.set_ylabel('Harga (AUD)')
plt.show()


# Penjelasan
print(f"Koefisien Korelasi: {corr}")
print(f"Berdasarkan Scatter Plot, Korelasi antara Nilai Harga Pangan Mata Uang Dolar dengan Nilai Harga Pangan Mata Uang AUD memiliki koefisien korelasi sebesar {corr} yang mendekati 1. Dengan nilai koefisien korelasi mendekati 1, maka korelasi antara Nilai Harga Pangan Mata Uang Dolar dengan Nilai Harga Pangan Mata Uang AUD berbanding lurus")
```
