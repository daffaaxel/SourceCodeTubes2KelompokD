```python

#Visualisasi Perbandingan Keterisian dan Kekosongan Terminal Charger dengan Hari
#Terisi
mon_isi = df["Mon"].sum()
tue_isi = df["Tues"].sum()
wed_isi = df["Wed"].sum()
thu_isi = df["Thurs"].sum()
fri_isi = df["Fri"].sum()
sat_isi = df["Sat"].sum()
sun_isi = df["Sun"].sum()
#Kosong
mon_song = df.loc[df["Mon"] == 0]
tue_song = df.loc[df["Tues"] == 0]
wed_song = df.loc[df["Wed"] == 0]
thu_song = df.loc[df["Thurs"] == 0]
fri_song = df.loc[df["Fri"] == 0]
sat_song = df.loc[df["Sat"] == 0]
sun_song = df.loc[df["Sun"] == 0]
#cari jumlah kosong
mon_songval = mon_song.value_counts().values
tue_songval = tue_song.value_counts().values
wed_songval = wed_song.value_counts().values
thu_songval = thu_song.value_counts().values
fri_songval = fri_song.value_counts().values
sat_songval = sat_song.value_counts().values
sun_songval = sun_song.value_counts().values

#Total kosong tiap harinya
mon_song_total = len(mon_songval)
tue_song_total = len(tue_songval)
wed_song_total = len(wed_songval)
thu_song_total = len(thu_songval)
fri_song_total = len(fri_songval)
sat_song_total = len(sat_songval)
sun_song_total = len(sun_songval)

data4 = {
    "Hari" : ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Terisi" : [mon_isi, tue_isi, wed_isi, thu_isi, fri_isi, sat_isi, fri_isi],
    "Kosong" : [mon_song_total, tue_song_total, wed_song_total, thu_song_total, fri_song_total, sat_song_total, sun_song_total]
}

df4 = pd.DataFrame(data4)
    
df4.plot(kind = "bar", x = "Hari", y = ["Terisi", "Kosong"], title = "Perbandingan Keterisian dan Kekosongan Terminal Charger dengan Hari")
```

```python
#Insight
#Berdasarkan visualisasi data tersebut, dapat dilihat bahwa secara umum keterisian terminal charger
#dalam seminggu tergolong rendah, dan terendah terjadi pada hari Sabtu. Keterisian tertinggi ada
#pada hari Kamis.
```

```python
#Visualisasi perbandingan platform dengan distance
#Sortir data sesuai hari
android = df.loc[(df["platform"] == "android")]
ios = df.loc[(df["platform"] == "ios")]
web = df.loc[(df["platform"] == "web")]

#Rata-rata
m_android = android["distance"].mean()
m_ios = ios["distance"].mean()
m_web = web["distance"].mean()

#Dataframe baru
data2 = {
    'plat' : ['android', 'ios', 'web'], 
    'Mean_distance' : [m_android, m_ios, m_web]
}
df2 = pd.DataFrame(data2)

#Visualisasi
color =  ["red", "green", "blue"]
df2.plot (kind = "barh", x = "plat", y = "Mean_distance", legend=True,title = "Perbandingan Jarak Tempuh(distance) dengan Platform Pengguna (Platform)", xlabel = "Rata-rata jarak tempuh", ylabel = "Platform", color = color)

```

```python
#Insight
#Berdasarkan visualisasi tersebut, diperoleh bahwa jarak tempuh terjauh dimiliki oleh pengguna
#dengan platform ios, sedangkan jarak tempuh terendah dimiliki oleh pengguna platform web
```

```python
#Perbandingan bulan terhadap rata-rata kwhTotal
#Membatasi data pada tahun 2015, bulan 1-5
a = df[13:88]

#Menyortir data tiap bulannya
jan = a[:8]
feb = a[8:17]
mar = a[17:40]
apr = a[40:56]
may = a[56:75]

#Mencari rata-rata kwhTotal tiap bulan
m_jan = jan["kwhTotal"].mean()
m_feb = feb["kwhTotal"].mean()
m_mar = mar["kwhTotal"].mean()
m_apr = apr["kwhTotal"].mean()
m_may = may["kwhTotal"].mean()

#Membuat data frame baru
data3 = {
    'Bulan' : ['jan', 'feb', 'mar', 'apr', 'may'],
    'Mean_kwhTotal' : [m_jan, m_feb, m_mar, m_apr, m_may]
}
df3 = pd.DataFrame(data3)

#Visualisasi
df3.plot(kind = "line", x = 'Bulan', y = 'Mean_kwhTotal', title = "Perbandingan bulan terhadap rata-rata kwhTotal")
```

```python
#Insight
#Berdasarkan visualisasi data, diperoleh bahwa paa bulan mei, rata-rata kwhTotal tertinggi, sedangkan
#kwhTotal terendah terjadi pada bulan April. Selain itu terjadi penurunan signifikan pada bulan
#Maret, yang awalnya pada bulan Februari hampir menyentuh angka 9, turun hingga menyentuh angka 7
```

```python
#Visualisasi Perbandingan Rata rata "chargeTimeHrs" dengan "Weekday" (Hari)
#Sortir data sesuai hari
mon = df.loc[(df["weekday"] == "Mon")]
tue = df.loc[(df["weekday"] == "Tue")]
wed = df.loc[(df["weekday"] == "Wed")]
thu = df.loc[(df["weekday"] == "Thu")]
fri = df.loc[(df["weekday"] == "Fri")]
sat = df.loc[(df["weekday"] == "Sat")]
sun = df.loc[(df["weekday"] == "Sun")]

#Cari rata-rata chargetime sesuai hari
m_mon = mon["chargeTimeHrs"].mean()
m_tue = tue["chargeTimeHrs"].mean()
m_wed = wed["chargeTimeHrs"].mean()
m_thu = thu["chargeTimeHrs"].mean()
m_fri = fri["chargeTimeHrs"].mean()
m_sat = sat["chargeTimeHrs"].mean()
m_sun = sun["chargeTimeHrs"].mean()

#Buat dataframe baru
data1 = {
    'Day' : ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Satruday', 'Sunday'], 
    'Mean_chargeTimeHrs' : [m_mon, m_tue, m_wed, m_thu, m_fri, m_sat, m_sun]
}
df1 = pd.DataFrame(data1)

#Visualisasi
color = ["red"]
df1.plot (kind = "line", x = "Day", y = "Mean_chargeTimeHrs", legend=True,title = "Perbandingan lama charge(chargeTimeHrs) dengan hari(weekdays)", xlabel = "Hari", ylabel = "Rata-rata lama charge", color = color)
```

```python
#Insight
#Berdasarkan visualisasi tersebut, terlihat bahwa terdapat tren penurunan rata-rata lama charge
#dengan hari, yang artinya semakin mendekati hari Minggu, rata-rata lama charge berkurang dan
#rata-rata lama charge terendah terjadi pada hari Minggu.
```

```python
#Korelasi kwh Total dengan harga
corr = df["kwhTotal"].corr(df["dollars"])
df.plot(kind="scatter", x = "dollars", y = "kwhTotal", xlabel = "Harga (dollar)", title = "Korelasi kwh Total dengan harga")

penjelasan = "" #string kosong
if corr == 0 :
    penjealasan = "Kedua nilai tidak berkorelasi"
elif corr > 0 :
    if corr == 1 :
        penjelasan = "Kedua nilai berkorelasi berbanding lurus"
    else : #0<korelasi<1
        penjelasan = "Kedua nilai cenderung berkorelasi berbanding lurus"
else : #Korelasi < 0
    if corr == -1 :
        penjelasan = "Kedua nilai berkorelasi berbanding terbalik"
    else : #-1<korelasi<0
        penjelasan = "Kedua nilai cenderung berkorelasi berbanding terbalik"
print (f"Nilai korelasi kwhTotal dan Harga(dollars) dari data adalah {corr}")
print (penjelasan)
```

```python
#Korelasi kwhTotal dengan lama charge (ChargetimeHrs)
corr = df["kwhTotal"].corr(df["chargeTimeHrs"])
df.plot(kind="scatter", x = "kwhTotal", y = "chargeTimeHrs", ylabel = "Lama charge", title = "Korelasi kwhTotal dengan lama charge (ChargetimeHrs)")

penjelasan = "" #string kosong
if corr == 0 :
    penjealasan = "Kedua nilai tidak berkorelasi"
elif corr > 0 :
    if corr == 1 :
        penjelasan = "Kedua nilai berkorelasi berbanding lurus"
    else : #0<korelasi<1
        penjelasan = "Kedua nilai cenderung berkorelasi berbanding lurus"
else : #Korelasi < 0
    if corr == -1 :
        penjelasan = "Kedua nilai berkorelasi berbanding terbalik"
    else : #-1<korelasi<0
        penjelasan = "Kedua nilai cenderung berkorelasi berbanding terbalik"
print (f"Nilai korelasi kwhTotal dan Lama Charge(Jam) dari data adalah {corr}")
print (penjelasan)
```
